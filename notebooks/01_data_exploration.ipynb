```python
import pandas as pd

#using a relative path
file_path = "../data/AmesHousing.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
#view the type of columns and datatypes
df.info()

# Quick descriptive statistics for numeric columns
df.describe().T
```

```python
# For non-numeric columns
cat_cols = df.select_dtypes(include=['object', 'category']).columns
print("Number of categorical columns:", len(cat_cols))
print(cat_cols.tolist())

for col in cat_cols:
    print(f"--- {col} ---")
    print(df[col].value_counts(dropna=False).head(10))  
    print("\n")
```

The most common observations of categorical variables include RL for MS Zoning, Pave for street, and NaN meaning null for Ally, thus meaning that Grvl is the only option for Alley and most observations do not have one.

```python
# outlier function
def find_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return series[(series < lower) | (series > upper)]

#checking salesprice
outliers = find_outliers_iqr(df['SalePrice'])
print(f"Outliers in SalePrice: {len(outliers)}")

# Remove outliers in SalePrice
Q1 = df['SalePrice'].quantile(0.25)
Q3 = df['SalePrice'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['SalePrice'] >= lower) & (df['SalePrice'] <= upper)]

print("Shape after outlier removal:", df.shape)
```

Correlation Modeling Section

```python

#only pull numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Compute Pearson correlation matrix
corr_matrix = numeric_df.corr(method='pearson')

# Check shape (should be around 30â€“40 columns)
print("Number of numeric columns:", len(numeric_df.columns))
corr_matrix.head()
```

Only the top 10 Correlated Variables

```python
corr_pairs = corr_matrix.unstack().reset_index()
corr_pairs.columns = ['Variable_1', 'Variable_2', 'Correlation']

# Drop self-correlations (a variable with itself)
corr_pairs = corr_pairs[corr_pairs['Variable_1'] != corr_pairs['Variable_2']]

# Keep only one of each pair (e.g., A-B but not B-A)
corr_pairs['Sorted'] = corr_pairs.apply(lambda x: tuple(sorted([x['Variable_1'], x['Variable_2']])), axis=1)
corr_pairs = corr_pairs.drop_duplicates(subset='Sorted').drop(columns='Sorted')

# Sort by absolute correlation
corr_pairs['AbsCorrelation'] = corr_pairs['Correlation'].abs()
top_corr = corr_pairs.sort_values(by='AbsCorrelation', ascending=False).head(10)

top_corr
```

We the strongest correlation is between Order and Yr Sold with a -97% R. Meaning a very strong negative linear relationship.

The dataset has 2,930 rows and 82 variables. Roughly 6.55% of the data is missing, mostly from Pool QC (99.6%), Misc Feature (96.4%), Alley (93.2%), and Fence (80.5%). Those fields rarely apply to most homes. After removing 137 outliers, 2,793 clean records remain.

Overall Quality, Gr Liv Area, and Garage Cars have no missing values. These three features strongly influence SalePrice, with correlations of 0.80, 0.71, and roughly 0.64, respectively. Year Built and Garage Yr Blt correlate at 0.83, showing that newer homes tend to have newer garages.
